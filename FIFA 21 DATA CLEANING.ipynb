```python
def inches_to_cm(val):
    if "'" in val:
        val = val.strip('"') #this removes the " symbiol at the end of all the inches value
        feet,inches = val.split("'") #this splits the values eg 5'11, so feet=5 and inches= 11
        inches_val = round((float(feet)*30.48 + float(inches)*2.54),2) #the calculation to convert ft and in to cm
        return inches_val
    else:
        val = val.strip('cm') #this removes the cm label on all cm values
        return val
    
fifa_data['Height'] = fifa_data['Height'].apply(lambda x: inches_to_cm(x)) #x is the input value for height values  
```

```python
# Lets clean the Value
# Lets handle the Value column
# Check if there are missing values

fifa_data.Value.isna().any()
fifa_data.Value.isnull().any()
```

```python
 #since there is no issue with the value then let work on the character
# Lets replace the currency first

fifa_data['Value'] = fifa_data['Value'].str.replace('€','')

# We will convert the M to million by multiplying by 1,000,000 and 1,000 in the case of K
new_values = []

for val in fifa_data.Value:
    if 'M' in val:
        raw_figure = float(val.split('M')[0])
        new_values.append(raw_figure*1000000)
    
    elif 'K' in val:
        raw_figure = float(val.split('K')[0])
        new_values.append(raw_figure*1000)
        
    else:
        new_values.append(val)

# Lets add it to the clean data
fifa_data['Value'] = pd.Series(new_values, dtype=float) 
```

```python
#let us handle the wage like we did for value
fifa_data['Wage'] = fifa_data['Wage'].str.replace('€','')
new_values = []

for val in fifa_data.Wage:
    if 'M' in val:
        raw_figure = float(val.split('M')[0])
        new_values.append(raw_figure*1000000)
    
    elif 'K' in val:
        raw_figure = float(val.split('K')[0])
        new_values.append(raw_figure*1000)
        
    else:
        new_values.append(val)
        
fifa_data['Wage'] = pd.Series(new_values, dtype=float)
```

```python
#let us handle the wage like we did for value
fifa_data['Release Clause'] = fifa_data['Release Clause'].str.replace('€','')
new_values = []

for val in fifa_data['Release Clause']:
    if 'M' in val:
        raw_figure = float(val.split('M')[0])
        new_values.append(raw_figure*1000000)
    
    elif 'K' in val:
        raw_figure = float(val.split('K')[0])
        new_values.append(raw_figure*1000)
        
    else:
        new_values.append(val)
        
fifa_data['Release Clause'] = pd.Series(new_values, dtype=float)
```
